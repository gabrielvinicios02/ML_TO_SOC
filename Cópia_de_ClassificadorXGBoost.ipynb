```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
#dataset = pd.read_csv('../content/drive/MyDrive/CICIDS2017/MachineLearningCVE/MachineLearningCVE.csv')
dataset = pd.read_csv('../content/drive/MyDrive/CICIDS2017/MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
#dataset = pd.read_csv('../content/drive/MyDrive/CICIDS2017/MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
#dataset = pd.read_csv('../content/drive/MyDrive/CICIDS2017/MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
#dataset = pd.read_csv('../content/drive/MyDrive/CICIDS2017/MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')

x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(dataset)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.2, random_state = 10)
```

```python
#para dados de treino
resultado = []

for elemento in y_train:
  if elemento == "DDoS":
    resultado.append(0)
  elif elemento == "BENIGN":
    resultado.append(1)

print(resultado)
```

```python
#para dados de teste
resultado2 = []

for elemento in y_test:
  if elemento == "DDoS":
    resultado2.append(0)
  elif elemento == "BENIGN":
    resultado2.append(1)

print(resultado2)
```

```python
print(y_train)
```

```python
y_train2 = np.array(resultado)
```

```python
y_test2 = np.array(resultado2)
```

```python
print(y_train2)
```

```python
print(y_test2)
```

```python
matriz = x_train

tem_nan = np.isnan(matriz)
tem_inf = np.isinf(matriz)

if np.any(tem_nan) or np.any(tem_inf):
    matriz_sem_nan = np.nan_to_num(matriz, nan=0, posinf=0, neginf=0)
    print("OcorrÃªncias de NaN encontradas. Matriz apÃ³s substituiÃ§Ã£o:")
    print(matriz_sem_nan)
else:
    print("NÃ£o foram encontradas ocorrÃªncias de NaN na matriz.")
```

```python
x_train2 = matriz_sem_nan
```

```python
from xgboost import XGBClassifier
Classifier = XGBClassifier()
Classifier.fit(x_train2, y_train2)
```

Matriz de confusÃ£o

```python
from sklearn.metrics import confusion_matrix, accuracy_score
ypred = Classifier.predict(x_test)
cm = confusion_matrix(y_test2, ypred)
print(cm)
```

PrecisÃ£o

```python
accuracy_score(y_test2, ypred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = Classifier, X = x_train2, y = y_train2, cv = 10)

print(accuracies.mean()*100)
print(accuracies.std()*100)
```
